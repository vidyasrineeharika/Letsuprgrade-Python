### Sub list


```python
lst=[1,1,6,4,2,5,3]
sublist = [1,1,5]
for i in range(len(lst)): 
    if lst[i] == sublist[0]:
        for j in range(i+1,len(lst)): 
            if lst[j] == sublist[1]:
                for k in range(j+1,len(lst)): 
                    if lst[k] == sublist[2]:
                        print("Its a Match") 
                        break
                break
        break
else:
    print("Its gone")
```
